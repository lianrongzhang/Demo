```python
def get_data(start_year,start_month,start_day,end_year,end_month,end_day): #
    start_date = str(date(start_year, start_month,start_day))
    end_date = str(date(start_year, end_month, end_day))
    date_list = pd.date_range(start_date, end_date, freq='D').strftime("%Y%m%d").tolist()

    column1 = ["報酬指數","收盤指數","漲跌(+/-)","漲跌點數","漲跌百分比(%)","特殊處理註記"] 
    column2 = ["成交統計","成交金額(元)","成交股數(股)","成交筆數"]
    column3 = ["類型","整體市場","股票"]
    column4 = ["證券代號","證券名稱","成交股數","成交筆數","成交金額","開盤價","最高價","最低價","收盤價","漲跌(+/-)","漲跌價差","最後揭示買價","最後揭示買量","最後揭示賣價","最後揭示賣量","本益比"]


    names = ['價格指數(臺灣證券交易所)','價格指數(跨市場)','價格指數(臺灣指數公司)','報酬指數(臺灣證券交易所)',
            '報酬指數(跨市場)','報酬指數(臺灣指數公司)','大盤統計資訊','漲跌證券數合計','每日收盤行情']
    
    for name in names:
        if not (os.path.exists('C:/Users/talen/Desktop/python/demo/'+str(name))):
            os.mkdir('C:/Users/talen/Desktop/python/demo/'+str(name))
        else:
            continue

    for Date in date_list:
        try:
            url = 'https://www.twse.com.tw/exchangeReport/MI_INDEX?response=json&date='+ Date +'&type=ALL'
            r = requests.post(url,headers=headers)
            df1 = pd.DataFrame(r.json()['data1'])
            df1.columns = column1
            df1.replace(["<p style ='color:red'>+</p>","<p style ='color:green'>-</p>"],['+','-'],inplace=True)
        
            df2 = pd.DataFrame(r.json()['data2'])
            df2.columns = column1
            df2.replace(["<p style ='color:red'>+</p>","<p style ='color:green'>-</p>"],['+','-'],inplace=True)
        
            df3 = pd.DataFrame(r.json()['data3'])
            df3.columns = column1
            df3.replace(["<p style ='color:red'>+</p>","<p style ='color:green'>-</p>"],['+','-'],inplace=True)

            df4 = pd.DataFrame(r.json()['data4'])
            df4.columns = column1
            df4.replace(["<p style ='color:red'>+</p>","<p style ='color:green'>-</p>"],['+','-'],inplace=True)
        
            df5 = pd.DataFrame(r.json()['data5'])
            df5.columns = column1
            df5.replace(["<p style ='color:red'>+</p>","<p style ='color:green'>-</p>"],['+','-'],inplace=True)

            df6 = pd.DataFrame(r.json()['data6'])
            df6.columns = column1
            df6.replace(["<p style ='color:red'>+</p>","<p style ='color:green'>-</p>"],['+','-'],inplace=True)

            df7 = pd.DataFrame(r.json()['data7'])
            df7.columns = column2

            df8 = pd.DataFrame(r.json()['data8'])
            df8.columns = column3

            df9 = pd.DataFrame(r.json()['data9'])
            df9.columns = column4
            df9.replace(['<p style= color:red>+</p>','<p style= color:green>-</p>','<p>X</p>','<p> </p>'],['+','-','--','--'],inplace=True)

            df1.to_csv("C:/Users/talen/Desktop/python/demo/價格指數(臺灣證券交易所)/價格指數(臺灣證券交易所)_{0}.csv".format(Date),encoding="utf_8_sig")
            df2.to_csv("C:/Users/talen/Desktop/python/demo/價格指數(跨市場)/價格指數(跨市場)_{0}.csv".format(Date),encoding="utf_8_sig")
            df3.to_csv("C:/Users/talen/Desktop/python/demo/價格指數(臺灣指數公司)/價格指數(臺灣指數公司)_{0}.csv".format(Date),encoding="utf_8_sig")
            df4.to_csv("C:/Users/talen/Desktop/python/demo/報酬指數(臺灣證券交易所)/報酬指數(臺灣證券交易所)_{0}.csv".format(Date),encoding="utf_8_sig")
            df5.to_csv("C:/Users/talen/Desktop/python/demo/報酬指數(跨市場)/報酬指數(跨市場)_{0}.csv".format(Date),encoding="utf_8_sig")
            df6.to_csv("C:/Users/talen/Desktop/python/demo/報酬指數(臺灣指數公司)/報酬指數(臺灣指數公司)_{0}.csv".format(Date),encoding="utf_8_sig")
            df7.to_csv("C:/Users/talen/Desktop/python/demo/大盤統計資訊/大盤統計資訊_{0}.csv".format(Date),encoding="utf_8_sig")
            df8.to_csv("C:/Users/talen/Desktop/python/demo/漲跌證券數合計/漲跌證券數合計_{0}.csv".format(Date),encoding="utf_8_sig")
            df9.to_csv("C:/Users/talen/Desktop/python/demo/每日收盤行情/每日收盤行情_{0}.csv".format(Date),encoding="utf_8_sig")
            
            sleep(delay)
        except:
            continue
    return "Done"
```
